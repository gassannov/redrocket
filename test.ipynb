```python
pixels_limit = 1920*1080
img1 = Image.open('test_pictures/bigimg.jpg')
w, h = img1.size
num_pixels =  w*h
```

```python
compress_koeff = 1
if num_pixels > pixels_limit:
    compress_koeff = pixels_limit/num_pixels
```

```python
import math
```

```python
resized_img =  img1.resize((int(w*math.sqrt(compress_koeff)), int(h*math.sqrt(compress_koeff))))
```

```python
resized_img.save('test_pictures/compr_image.jpg')
```
